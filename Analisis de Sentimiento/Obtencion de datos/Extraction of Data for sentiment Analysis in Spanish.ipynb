```python
scraper = Nitter(0, skip_instance_check = False)

columns = ['user', 'text', 'date', 'emotion', 'sentiment', ]



sentiment_dataframe = pd.DataFrame(columns=columns)
```

```python
if __name__ == "__main__":
    
    filters = ['videos',  'native_video', 'replies', 'links', 'images', 'pro_video', 'media', 'nativeretweets', 'news']

    number_of_tweets = 10


    for sentiment in sentiments:
        for emotion in sentiments[sentiment]:
            print("Emotion: ", emotion)


            words= sentiments[sentiment][emotion][0:15]
            recopilation = []
            for word in words:
                try:
                    tweets = scraper.get_tweets(word, mode = 'term',number = number_of_tweets, exclude = filters, language = 'es')
                
                except:                    
                    print("="*10)
                    print('VETEALV')
                    print("Sentiment: ", sentiment)
                    print("word: ", word)
                    print("="*10)
                    continue

                
                for tweet in tweets['tweets']:
                    data = [tweet['user']['username'], tweet['text'],tweet['date'], emotion, sentiment]
                    recopilation.append(data)


            dataframe = pd.DataFrame(recopilation, columns=columns)
            dataframe.to_csv(f'./DataFrames/{sentiment}_{emotion}.csv', index=False)

            sentiment_dataframe = pd.concat([dataframe, sentiment_dataframe], ignore_index = True)


    sentiment_dataframe.to_csv('./sentiment_dataframe.csv', index=False)
```
